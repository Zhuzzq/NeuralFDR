```r
source("https://bioconductor.org/biocLite.R")
biocLite("IHWpaper")
```

```r
library("IHW")
library("IHWpaper")
```

```r
sim <- du_ttest_sim(80000, 0.9, 2, seed=1)
```

```r
ihw_res <- ihw(sim$pvalue, sim$filterstat, alpha = 0.05, nbins = 20)
```

```r
rejections(ihw_res)
```

```r
plot(ihw_res)
```

```r
write.csv(sim,'sim.csv')
```
