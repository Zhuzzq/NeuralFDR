```r
data = read.csv('../data/2d_slope_for_ihw.csv', head=TRUE)
```

```r
library("IHW")
```

```r
nalpha = 0.05
```

```r
data$x_value = as.factor(data$x_value)
```

```r
ihwRes <- ihw(p_value ~ x_value ,  data = data, alpha = nalpha)
```

```r
rejections(ihwRes)
```

```r
rej <- adj_pvalues(ihwRes) < nalpha
```

```r
trej <- rej * data$h
```

```r
1 - sum(trej) / sum(rej)
```

```r
plot(ihwRes)
```

```r
w = weights(ihwRes, levels_only = TRUE)
```

```r
write.csv(w, file = 'ihw2dgmweights2.csv')
```
