```r
data = read.csv('../data/data_5d_gm_ihw.csv', head=TRUE)
```

```r
library("IHW")
```

```r
nalpha = 0.05
```

```r
data$group = as.factor(data$group)
```

```r
ihwRes <- ihw(p_value ~ group ,  data = data, alpha = nalpha)
```

```r
rejections(ihwRes)
```

```r
rej <- adj_pvalues(ihwRes) < nalpha
```

```r
trej <- rej * data$h
```

```r
1 - sum(trej) / sum(rej)
```

```r
plot(ihwRes)
```

```r
w = weights(ihwRes, levels_only = TRUE)
```

```r
write.csv(w, file = 'ihw2dgmweights2.csv')
```
