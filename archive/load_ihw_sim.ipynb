```python
from sideinfo import *
%load_ext autoreload
%autoreload 2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd

```

```python
for i in range(1,11):
    df = pd.read_csv('sim.csv. '+str(i))
    h = np.array(df.H)
    x = np.array(df.filterstat)
    p = np.array(df.pvalue)
    with open('../data_rep/data_ihw_sim.csv.' + str(i), 'w') as f:
        f.write('x_value, p_value, h\n')
        for i in range(len(x)):
            f.write("{}, {}, {}\n".format(x[i],p[i], h[i]))
```
